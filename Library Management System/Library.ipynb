```python
import pandas as pd

data = pd.read_csv("library.csv")


total_books = data.shape[0]
total_issued_books = data["Issued_Count"].sum()

most_issued_book = data.loc[data["Issued_Count"].idxmax(), "Book_Name"]
min_available_book = data.loc[
    (data["Total_Quantity"] - data["Issued_Count"]).idxmin(),
    "Book_Name"
]


total_students = total_issued_books   

print("LIBRARY SUMMARY REPORT")
print(f"Total Books              : {total_books}")
print(f"Total Issued Books       : {total_issued_books}")
print(f"Total Students Issued    : {total_students}")
print(f"Most Popular Book        : {most_issued_book}")
print(f"Least Available Book     : {min_available_book}")
```
