```python
import pandas as pd

data = pd.read_csv("students.csv")

print("Student Dataset:")
print(data)

data["Average"] = data[["Subject1", "Subject2", "Subject3"]].mean(axis=1)

highest_marks = data[["Subject1", "Subject2", "Subject3"]].max().max()
lowest_marks = data[["Subject1", "Subject2", "Subject3"]].min().min()

top_student = data.loc[data["Average"].idxmax()]

print("\n Analysis Results:")
print(data)

print("\n Highest Marks:", highest_marks)
print(" Lowest Marks:", lowest_marks)

print("\n Top Scoring Student:")
print("Name:", top_student["Student Name"])
print("Average Marks:", top_student["Average"])
```
