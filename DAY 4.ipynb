```python

import numpy as np


array_1d = np.array([10, 20, 30, 40, 50])

array_2d = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("1D Array:", array_1d)
print("2D Array:\n", array_2d)

addition = array_1d + 5
multiplication = array_1d * 2
max_value = np.max(array_1d)

print("\nAfter Addition (+5):", addition)
print("After Multiplication (*2):", multiplication)
print("Maximum Value:", max_value)
```
